```python
from ultralytics import YOLO
model = YOLO("yolov8n.pt")  # завантажиться автоматично
results = model("https://ultralytics.com/images/bus.jpg")

r = results[0]
print(r.boxes.xyxy)   # координати боксів
print(r.boxes.conf)   # впевненість
print(r.boxes.cls)    # індекси класів
r.show()              # картинка з боксами
```

```python
results = model("https://ultralytics.com/images/bus.jpg", conf=0.01)
print(len(results[0].boxes))  # порівняй з тим, що було при дефолті
```
